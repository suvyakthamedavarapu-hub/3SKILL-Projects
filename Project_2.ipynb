TASK - 4: Text Feature Engineering

Step - 1: Combine the Skills, Certifications and Job Role into one column:
combined_text

```python
df['combined_text'] = (
    df['Skills'] + ' ' +
    df['Certifications'] + ' ' +
    df['Job Role']
)
```

```python
df['combined_text'].head()
```

Step - 2: Clean the text:

• Convert to lowercase

• Remove special characters

• Remove extra spaces


```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z ]', '', text)
    text = re.sub(r'\s+', ' ', text)
    return text

df['combined_text'] = df['combined_text'].apply(clean_text)
```

```python
df['combined_text'].head()
```

Step - 3: [Q & A]

Question: Explain why text cleaning is necessary before vectorization.

Answer: Text cleaning is necessary because raw text may contain noise and inconsistencies, which can negatively impact machine learning models.

This preprocessing ensures consistency and improves the effectiveness of text vectorization.

TASK 5: Convert Text to Numerical Features

Step - 1: Use TfidfVectorizer


```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=100)

text_features = tfidf.fit_transform(df['combined_text']).toarray()
```

```python
text_features.shape
```

Step - 2: [Q & A] Explain:

Question: Why ML models cannot work directly with text.

Answer: Text data cannot be directly used by machine learning models, as they require numerical input. Therefore, the combined_text feature was converted into numerical form using TF-IDF

Question: What TF-IDF represents.

Answer: TF-IDF assigns importance to words based on how frequently they appear in a document and how unique they are across the dataset. This helps in capturing meaningful textual patterns for model training.

Task - 10: Model Training

Train at least four models:
1. Logistic Regression
2. Random Forest
3. Support Vector Machine
4. K-Nearest Neighbors

Use scaled/unscaled data appropriately.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(),
    "KNN": KNeighborsClassifier()
}

for name, model in models.items():
    model.fit(X_train, y_train)
    pred = model.predict(X_test)

    from sklearn.metrics import accuracy_score
    print(f"{name} Accuracy:", accuracy_score(y_test, pred))
```

TASK 11: Model Evaluation and Comparison

Step - 1: Print accuracy score for all models.

Step - 2: Display classification report.

```python
from sklearn.metrics import classification_report, accuracy_score

results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    pred = model.predict(X_test)

    acc = accuracy_score(y_test, pred)
    results[name] = acc

    print(f"\n{name}")
    print("Accuracy:", acc)
    print(classification_report(y_test, pred))
```

Task - 13: Build Hiring Prediction Function

Create a function that:

• Takes user input:
• Skills
• Experience
• Education
• Certifications
• Projects Count

• Salary Expectation
• Returns:
• Prediction (Hire or Reject)
• Probability score

This simulates a real AI resume screening tool.

```python
def predict_hiring(skills, experience, education, certifications, projects, salary):

    # Step 1: Combine text
    text = skills + " " + certifications + " " + "job"

    # Step 2: Clean text
    import re
    text = text.lower()
    text = re.sub(r'[^a-zA-Z ]', '', text)
    text = re.sub(r'\s+', ' ', text)

    # Step 3: TF-IDF transform
    text_vec = tfidf.transform([text]).toarray()

    # Step 4: Encode education
    edu_val = le.transform([education])[0]

    # Step 5: Numeric features
    num = [[experience, salary, projects, edu_val]]

    # Step 6: Combine
    import numpy as np
    final_input = np.hstack((text_vec, num))

    # Step 7: Scale
    final_input = scaler.transform(final_input)

    # Step 8: Predict
    actual_best_model = models[best_model]
    pred = actual_best_model.predict(final_input)
    prob = actual_best_model.predict_proba(final_input)

    # Step 9: Output
    result = "Hire" if pred[0] == 1 else "Reject"

    return result, prob[0][1]
```

Testing the function:

To check whether the model predicts correctly or not?

```python
predict_hiring(
    skills="python machine learning data analysis",
    experience=3,
    education="B.Tech",
    certifications="aws certification",
    projects=5,
    salary=70000
)
```
